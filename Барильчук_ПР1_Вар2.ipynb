```python
import numpy as np
# 1. Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
arr = np.arange(1, 21)
print(arr)
```

```python
# 2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його випадковими натуральними числами та виведіть його значення.
mat = np.random.randint(1, 101, size=(3, 3))
print(mat)
```

```python
# 3. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
arr5x5 = np.random.uniform(-1, 1, (5, 5))
print(arr5x5)
```

```python
# 4. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть середнє значення елементів.
arr = np.random.rand(5, 5)
print("Масив:\n", arr)
mean = np.mean(arr)
print("\nСереднє значення:", mean)
```

```python
# 5. Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими
# числами в діапазоні від -10 до 10 та виконайте на них поелементні операції додавання, віднімання та множення.
arr1 = np.random.randint(-10, 11, 5)
arr2 = np.random.randint(-10, 11, 5)
print("Перший масив:", arr1)
print("Другий масив:", arr2)
sum_arr = arr1 + arr2
diff_arr = arr1 - arr2
prod_arr = arr1 * arr2
print("\nДодавання:", sum_arr)
print("Віднімання:", diff_arr)
print("Множення:", prod_arr)
```

```python
# 6. Створіть дві матриці 2x2, заповніть їх довільними дійсними числами від 3 до 10 та знайдіть їх добуток.
mat1 = np.random.uniform(3, 10, (2, 2))
mat2 = np.random.uniform(3, 10, (2, 2))
print("Перша матриця:\n", mat1)
print("Друга матриця:\n", mat2)
prod_mat = np.dot(mat1, mat2)
print("\nДобуток матриць:\n", prod_mat)
```

```python
# 7. Створіть дві матриці розміром 2x3 та 3x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

mat1 = np.random.randint(1, 6, (2, 3))
mat2 = np.random.randint(1, 6, (3, 2))
print("Перша матриця (2x3):\n", mat1)
print("Друга матриця (3x2):\n", mat2)
prod_mat = np.dot(mat1, mat2)
print("\nДобуток матриць:\n", prod_mat)
```

```python
# 8. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.
mat = np.random.randint(1, 13, (3, 3))
print("Матриця 3x3:\n", mat)
if np.linalg.det(mat) != 0:
    inv_mat = np.linalg.inv(mat)
    print("\nОбернена матриця:\n", inv_mat)
else:
    print("\nМатриця не має оберненої матриці.")
```

```python
# 9. Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -8 до 8 та транспонуйте її.
mat = np.random.uniform(-8, 8, (4, 4))
print("Матриця 4x4:\n", mat)
transp_mat = mat.T
print("\nТранспонована матриця:\n", transp_mat)
```

```python
# 10. Створіть матрицю розміром 3x4 та вектор розміром 4 x1, заповніть їх випадковими
#цілими числами в діапазоні від 1 до 10 та знайдіть середнє значення елементів матриці.
mat = np.random.randint(1, 11, (3, 4))
vect = np.random.randint(1, 11, (4, 1))
print("Матриця 3x4:\n", mat)
print("\nВектор 4x1:\n", vect)
mean = np.mean(mat)
print("\nСереднє значення елементів матриці:", mean)
```

```python
# 11. Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю
#випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними
#числами від 0 до 1, і знайдіть добуток матриці на вектор.
mat = np.random.randint(1, 11, (2, 3))
vect = np.random.uniform(0, 1, (3, 1))
print("Матриця 2x3:\n", mat)
print("\nВектор 3x1:\n", vect)
prod = np.dot(mat, vect)
print("\nДобуток матриці на вектор:\n", prod)
```

```python
# 12. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
mat1 = np.random.randint(1, 11, (3, 3))
mat2 = np.random.randint(1, 11, (3, 3))
print("Перша матриця 3x3:\n", mat1)
print("\nДруга матриця 3x3:\n", mat2)
diff_mat = mat1 - mat2
print("\nРізниця матриць:\n", diff_mat)
```

```python
# 13. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
mat1 = np.random.randint(-5, 6, (3, 3))
mat2 = np.random.randint(-5, 6, (3, 3))
print("Перша матриця 3x3:\n", mat1)
print("\nДруга матриця 3x3:\n", mat2)
prod_mat = np.dot(mat1, mat2)
print("\nДобуток матриць:\n", prod_mat)
```

```python
# 14. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від -5 до 5 та знайдіть суму елементів матриці.
mat = np.random.randint(-5, 6, (3, 3))
print("Матриця 3x3:\n", mat)
sum = np.sum(mat)
print("\nСума елементів матриці:", sum)
```

```python
# 15. Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в
#діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить добуток елементів кожного рядка матриці.
mat = np.random.uniform(0, 1, (4, 4))
print("Матриця 4x4:\n", mat)
row_prods = np.prod(mat, axis=1)
col_vect = row_prods.reshape(-1, 1)
print("\nВектор-стовпчик з добутками елементів кожного рядка:\n", col_vect)
```
