```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_diabetes
import pandas as pd
```

```python
data = load_diabetes()

df = pd.DataFrame(data=data.data, columns=data.feature_names)

# Виведення перших рядків
print("Перші рядки даних:")
df.head()
```

```python
#Перевірка на пропущені значення
missing_values = df.isnull().sum()
#Перевірка наявності дублікатів
duplicates = df.duplicated().sum()
# Виведення результатів
print("Пропущені значення в кожній колонці:\n", missing_values)
print("\nКількість дублікатів:", duplicates)
```

```python
#Перевірити тип даних. При необхідності замінити тип даних на числовий
df.dtypes
```

```python
# Обчислити кореляцію між прогресом діабету і іншмим ознаками.
#Розташувати в порядку спадання.
#Побудувати теплову карту кореляції.
df['target'] = data.target
correlation = df.corr()['target'].drop('target')
sorted_correlation = correlation.sort_values(ascending=False)
print("Кореляція між прогресом діабету і іншими ознаками, в порядку спадання:")
print(sorted_correlation)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Теплова карта кореляції ознак з прогресом діабету')
plt.show()
```

```python
# Масштабувати ознаки.
features = df.drop('target', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df['target'] = df['target']
print("Перші рядки з масштабованими ознаками:")
scaled_df.head()
```

```python
#Поділити дані на тренувальну і тестову вибірки.
X_train, X_test, y_train, y_test = train_test_split(scaled_features, scaled_df, test_size=0.2, random_state=42)
print("Розмір тренувальної вибірки:", X_train.shape)
print("Розмір тестової вибірки:", X_test.shape)
```

```python
#Побудувати модель лінійной регресії і RandomForest регресії на
#основі більш корелюючих ознак (підібрати оптимальне число).
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred_lin = lin_reg.predict(X_test)

# Вивести коефіцієнт детермінації і середньоквадратичну помилку.
mse_lin = mean_squared_error(y_test, y_pred_lin)
r2_lin = r2_score(y_test, y_pred_lin)

#Побудова Random Forest регресії
rf_reg = RandomForestRegressor(random_state=42, n_estimators=100)
rf_reg.fit(X_train, y_train)
y_pred_rf = rf_reg.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Лінійна регресія: MSE =", mse_lin, ", R2 =", r2_lin)
print("Random Forest регресія: MSE =", mse_rf, ", R2 =", r2_rf)
```

```python
#Вивести справжні і прогнозовані значення.
model1 = LinearRegression()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
results = pd.DataFrame({
    'Справжні': y_test.to_numpy().ravel(),
    'Прогнозовані': y_pred.ravel()})
results = results.sort_values(by='Прогнозовані', ascending=False)
print("\nSorted Results:")
print(results)
```

```python
# Побудувати графіки.
y_test_1d = y_test.to_numpy().ravel()
y_pred_lin_1d = y_pred_lin.ravel()
y_pred_rf_1d = y_pred_rf.ravel()
plt.figure(figsize=(14, 6))

# Scatter Plot для Лінійної регресії
plt.subplot(1, 2, 1)
plt.scatter(y_test_1d, y_pred_lin_1d, color='blue', alpha=0.6, edgecolor='k')
plt.plot([y_test_1d.min(), y_test_1d.max()], [y_test_1d.min(), y_test_1d.max()], color='red', linewidth=2)
plt.title('Лінійна регресія')
plt.xlabel('Справжні значення')
plt.ylabel('Прогнозовані значення')
plt.grid(True)

# Scatter Plot для Random Forest регресії
plt.subplot(1, 2, 2)
plt.scatter(y_test_1d, y_pred_rf_1d, color='green', alpha=0.6, edgecolor='k')
plt.plot([y_test_1d.min(), y_test_1d.max()], [y_test_1d.min(), y_test_1d.max()], color='red', linewidth=2)
plt.title('Random Forest регресія')
plt.xlabel('Справжні значення')
plt.ylabel('Прогнозовані значення')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Висновок:**
У цьому завданні я виконав аналіз даних з набору `diabetes` і обчислив кореляцію між ознаками та прогресом діабету, що дозволило мені визначити найбільш впливові фактори. Використавши ці ознаки, побудував моделі лінійної регресії та Random Forest регресії, підібравши оптимальне число ознак для досягнення найкращих результатів. Після масштабування ознак і поділу на тренувальну та тестову вибірки я оцінив ефективність моделей за допомогою коефіцієнта детермінації (R²) і середньоквадратичної помилки (MSE). Справжні та прогнозовані значення для обох моделей показали, що регресія Random Forest демонструє вищу точність порівняно з лінійною регресією, що підтверджується кращими значеннями R² і MSE. Це свідчить про те, що більш складна модель краще враховує нелінійні залежності в даних.

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()

df = pd.DataFrame(data=data.data, columns=data.feature_names)

# Виведення перших рядків
print("Перші рядки даних:")
df.head()
```

```python
#Перевірка на пропущені значення
missing_values = df.isnull().sum()
#Перевірка наявності дублікатів
duplicates = df.duplicated().sum()
# Виведення результатів
print("Пропущені значення в кожній колонці:\n", missing_values)
print("\nКількість дублікатів:", duplicates)
```

```python
#Перевірити тип даних. При необхідності замінити тип даних на числовий
df.dtypes
```

```python
# Обчислити кореляцію між ознаками.
#Розташувати в порядку спадання.
#Побудувати теплову карту кореляції.
df['target'] = data.target
correlation = df.corr()['target'].drop('target')
sorted_correlation = correlation.sort_values(ascending=False)
print("Кореляція в порядку спадання:")
print(sorted_correlation)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Теплова карта кореляції ознак з прогресом діабету')
plt.show()
```

```python
# Масштабувати ознаки.
features = df.drop('target', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df['target'] = df['target']
print("Перші рядки з масштабованими ознаками:")
scaled_df.head()
```

```python
#Поділити дані на тренувальну і тестову вибірки.
X_train, X_test, y_train, y_test = train_test_split(scaled_features, scaled_df, test_size=0.2, random_state=42)
print("Розмір тренувальної вибірки:", X_train.shape)
print("Розмір тестової вибірки:", X_test.shape)
```

```python
#Побудувати 3 моделі: лін. Регресія, регресія випадковий ліс, Ridge.
#Підібрати оптимальні параметри за допомогою GridSeach.
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split, GridSearchCV

# Лінійна регресія
lin_reg = LinearRegression()

# Random Forest регресія
rf_reg = RandomForestRegressor(random_state=42)

# Ridge-регресія
ridge_reg = Ridge()

# Набір параметрів для пошуку
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

param_grid_ridge = {
    'alpha': [0.1, 1.0, 10.0, 100.0],
    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
}

# Використання GridSearchCV для підбору параметрів
grid_search_rf = GridSearchCV(rf_reg, param_grid_rf, cv=5, scoring='neg_mean_squared_error')
grid_search_ridge = GridSearchCV(ridge_reg, param_grid_ridge, cv=5, scoring='neg_mean_squared_error')

# Навчання моделей
lin_reg.fit(X_train, y_train)

grid_search_rf.fit(X_train, y_train)
grid_search_ridge.fit(X_train, y_train)

# Оптимальні параметри
best_rf = grid_search_rf.best_estimator_
best_ridge = grid_search_ridge.best_estimator_

# Прогнозування
y_pred_lin = lin_reg.predict(X_test)
y_pred_rf = best_rf.predict(X_test)
y_pred_ridge = best_ridge.predict(X_test)

# Обчислення метрик
r2_lin = r2_score(y_test, y_pred_lin)
mse_lin = mean_squared_error(y_test, y_pred_lin)

r2_rf = r2_score(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)

r2_ridge = r2_score(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)

# Виведення результатів
print(f"Лінійна регресія - R²: {r2_lin:.4f}, MSE: {mse_lin:.4f}")
print(f"Random Forest регресія - R²: {r2_rf:.4f}, MSE: {mse_rf:.4f}")
print(f"Ridge регресія - R²: {r2_ridge:.4f}, MSE: {mse_ridge:.4f}")

print("\nОптимальні параметри для Random Forest:", grid_search_rf.best_params_)
print("Оптимальні параметри для Ridge регресії:", grid_search_ridge.best_params_)
```

```python
#Побудувати графіки
plt.figure(figsize=(18, 6))

# Scatter Plot для Лінійної регресії
plt.subplot(1, 3, 1)
plt.scatter(y_test, y_pred_lin, color='blue', alpha=0.6, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.title('Лінійна регресія')
plt.xlabel('Справжні значення')
plt.ylabel('Прогнозовані значення')
plt.grid(True)

# Scatter Plot для Random Forest регресії
plt.subplot(1, 3, 2)
plt.scatter(y_test, y_pred_rf, color='green', alpha=0.6, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.title('Random Forest регресія')
plt.xlabel('Справжні значення')
plt.ylabel('Прогнозовані значення')
plt.grid(True)

# Scatter Plot для Ridge регресії
plt.subplot(1, 3, 3)
plt.scatter(y_test, y_pred_ridge, color='purple', alpha=0.6, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.title('Ridge регресія')
plt.xlabel('Справжні значення')
plt.ylabel('Прогнозовані значення')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Виведення справжніх і прогнозованих значень
results_df = pd.DataFrame({
    'Справжні значення': np.array(y_test).flatten(),
    'Лінійна регресія (прогноз)': np.array(y_pred_lin).flatten(),
    'Random Forest регресія (прогноз)': np.array(y_pred_rf).flatten(),
    'Ridge регресія (прогноз)': np.array(y_pred_ridge).flatten()
})

# Виведення перших 10 значень
print(results_df.head(10))
```

Для кожної моделі були підібрані оптимальні параметри за допомогою GridSearchCV, після чого було проведено оцінку ефективності моделей за метриками R² та середньоквадратичної помилки.

Це свідчить про те, що модель Random Forest краще вміє знаходити складні нелінійні залежності в даних, що дозволяє їй точніше прогнозувати значення цільової змінної у порівнянні з лінійною та Ridge регресіями.
